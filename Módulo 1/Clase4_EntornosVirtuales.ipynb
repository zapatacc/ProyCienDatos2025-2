# ğŸ“‘ GestiÃ³n de dependencias con `requirements.txt`

El archivo `requirements.txt` es la forma **clÃ¡sica y mÃ¡s simple** de manejar dependencias en proyectos de Python.

Sirve para **listar las librerÃ­as necesarias** para ejecutar un proyecto y, opcionalmente, fijar versiones especÃ­ficas.

El archivo `requirements.txt` es un documento que lista todas las bibliotecas y paquetes de Python que tu proyecto necesita para funcionar correctamente. Es especialmente Ãºtil cuando quieres compartir tu proyecto con otras personas o cuando lo quieres desplegar en otro entorno (por ejemplo, en un servidor).

# ğŸš€ Manejadores modernos de dependencias

Ya vimos que `requirements.txt` es la forma **clÃ¡sica** de gestionar dependencias en Python.

- No administra entornos virtuales por sÃ­ mismo.

- Es difÃ­cil manejar dependencias de desarrollo vs dependencias de producciÃ³n.
- No integra metadatos del proyecto (nombre, versiÃ³n, autor, etc.).

ğŸ‘‰ Por eso surgieron herramientas **modernas de gestiÃ³n de dependencias** que extienden y mejoran el flujo de trabajo.

## ğŸ”¹ Ventajas de los manejadores modernos

- Algunos permiten **publicar paquetes** en PyPI fÃ¡cilmente (ej. Poetry, Hatch).

---

## ğŸ”¹ Principales herramientas

# ğŸ“‚ Archivos de configuraciÃ³n y bloqueo

Hasta ahora vimos cÃ³mo manejar dependencias con `requirements.txt`.
Sin embargo, los manejadores modernos de dependencias utilizan un **par de archivos clave** que mejoran la reproducibilidad y organizaciÃ³n:

---

## 1. `pyproject.toml`
ğŸ“Œ Es un archivo de configuraciÃ³n estÃ¡ndar en Python (definido por [PEP 518](https://peps.python.org/pep-0518/))
- Contiene la informaciÃ³n del proyecto: nombre, versiÃ³n, autor, dependencias, build system, etc.
- Sirve como **fuente Ãºnica de la verdad** para configurar tu entorno.
- Diferentes herramientas lo leen: `poetry`, `uv`, `hatch`, `pdm`, etc.

ğŸ‘‰ Ejemplo bÃ¡sico de `pyproject.toml` con `poetry`:
```toml
[tool.poetry]
name = "mi-proyecto"
version = "0.1.0"
description = "Ejemplo de uso de Poetry"
authors = ["Cristian Zapata <cristianz@iteso.mx>"]

[tool.poetry.dependencies]
python = "^3.11"
numpy = "^1.26"
pandas = "^2.0"
```

---

## 2. `poetry.lock` / `uv.lock` / `pdm.lock`
ğŸ“Œ Estos son **archivos de bloqueo de dependencias**.
- Se generan automÃ¡ticamente cuando instalas dependencias.
- Guardan las **versiones exactas** que se resolvieron en la instalaciÃ³n.
- Garantizan que tu proyecto se pueda reproducir en cualquier mÃ¡quina de forma idÃ©ntica.

ğŸ‘‰ Ejemplo:
- En un equipo se guarda que `numpy==1.26.4` y `pandas==2.1.0`.
- En otro equipo, al usar `poetry install` o `uv sync`, instalarÃ¡ **exactamente esas versiones**, evitando inconsistencias.

---

âœ… Diferencia con `requirements.txt`:
- `requirements.txt` lo defines tÃº manualmente.
- `.lock` lo genera la herramienta automÃ¡ticamente y siempre refleja **el estado real del entorno**.

---

ğŸ“Œ En conclusiÃ³n:
- `pyproject.toml` = **definiciÃ³n de lo que quieres**.
- `.lock` = **fotografÃ­a exacta de lo que tienes instalado**.

# âš¡ Uso bÃ¡sico de `uv`: entornos, dependencias y reproducibilidad

`uv` es un gestor ultrarrÃ¡pido que simplifica el ciclo: **crear entorno â†’ declarar dependencias â†’ ejecutar â†’ reproducir**.

## 1) Crear el entorno virtual

> Desde la carpeta raÃ­z del proyecto (que ya tiene tu cÃ³digo y estructura).

```bash
# Crea un entorno virtual en .venv/ (recomendado)
uv venv
```

- Crea `.venv/` con la versiÃ³n de Python por defecto del sistema (o la pinnada si existe `.python-version`).

## 2) Agregar dependencias

```bash
# Agregar dependencias de ejecuciÃ³n
uv add numpy pandas

# Agregar dependencias de desarrollo (por ejemplo, linters/test)
uv add --dev pytest black
```

- Crea `.venv/` si no existe.
- Instala **exactamente** las versiones fijadas en `uv.lock`.

`uv run` habilita el entorno sÃ³lo para ese comando.
